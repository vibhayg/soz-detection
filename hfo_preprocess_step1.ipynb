```python
file_names = []
for i in os.listdir('../nd/'):
    file_names.append(i)
file_names.sort()
file_names
```

```python
list = []
for i in file_names:
    f = pd.read_csv(f'../nd/{i}', sep = '\t')
    x = len(f.index)
    list.append(x)
sum(list)
list


    
```

```python
data_frames = []
for file_name in file_names:
    temp = pd.read_csv(f'../nd/{file_name}', sep = '\t')
    data_frames.append(temp)

data_frames[29] 

for df in data_frames:
    rate = []
    for i in range(len(df.index)):
        x = (df.at[i, 'fmax_FR']/df.at[i, 'indDuration'])
        rate.append(x)
    #df.shape
    #len(rate)
    df.insert(15,'rate', rate, True)
  

data_frames[2].columns


```

```python
new_data_frames = []
for df in data_frames:
    df_sub = df.drop(df.columns.difference(['strChannelName', 'Amplpp', 'PowmaxFR', 'rate']), 1)
    new_data_frames.append(df_sub)
    
    
```

```python
total = []
for n in new_data_frames:
    for i in range(len(n.index)):
        p = n.at[i, 'strChannelName']
        if p not in total:
            total.append(p)
total.sort()
        
total



```

```python

columns_1 = []
for i in total:
    columns_1.append(i + '.Amplpp')
    columns_1.append(i + '.PowmaxFR')
    columns_1.append(i + '.rate')

empty_list = []
for i in range(52*3):
    empty_list.append(0)
empty_zeroes = []
for i in range(30):
    empty_zeroes.append(empty_list)

data = empty_zeroes
final_data_frame = pd.DataFrame(data, columns = columns_1)

final_data_frame.shape

user = 0
for df in new_data_frames:
    m = []
    elec = 0
    for k in total:
        x = 0
        y = 0
        z = 0
        for i in range(len(df.index)):
            if df.at[i,'strChannelName'] == k:
                x = max(x, df.at[i, 'Amplpp'])
                y = max(y, df.at[i, 'PowmaxFR'])
                z = max(z, df.at[i, 'rate'])
        #print(elec)
        #print(user, columns_1[elec * 3 + 0])
        final_data_frame.at[user, columns_1[elec * 3 + 0]] = x
        final_data_frame.at[user, columns_1[elec * 3 + 1]] = y
        final_data_frame.at[user, columns_1[elec * 3 + 2]] = z
        elec += 1
    user += 1    
            
final_data_frame
```
