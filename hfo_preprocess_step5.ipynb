```python
file_names = []
data_frames = []
for root, dirs, files in os.walk("../ds003498-download/"):
    for file in files:
        if file.endswith(".tsv"):
             path = os.path.join(root, file)
             if 'events.tsv' in path:  
                file_names.append(path)
                    
file_names.sort()
file_names
len(file_names)

for file in file_names:
    df = pd.read_csv(file, sep = '\t')
    data_frames.append(df)
```

```python
list_of_electrodes = []
for df in data_frames:
    for j in list(df.iloc[:,2]):
        x = j.find('_')
        j_1 = j[x+1:]
        if j_1 not in list_of_electrodes:
            list_of_electrodes.append(j_1)
```
