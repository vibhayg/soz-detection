```python
#goal - create a dataframe with:
    #number of rows should be 4 * numbers of runs for each subject
    #variables are average duration of HFO events
        #HFO rate
        #brain areas

#2,4,5,6,7,8,(9)
#lists all filenames in the directory that end with tsv extension in list file_names
file_names = []
data_frames = []
for root, dirs, files in os.walk("../ds003498-download/"):
    for file in files:
        if file.endswith(".tsv"):
             path = os.path.join(root, file)
             if (('sub-02' in path) or ('sub-04' in path) or ('sub-05' in path) or ('sub-06' in path) or ('sub-07' in path) or ('sub-08' in path) or ('sub-09' in path)) and 'events.tsv' in path:  
                file_names.append(path)
                    
file_names.sort()
file_names
len(file_names)

for file in file_names:
    df = pd.read_csv(file, sep = '\t')
    data_frames.append(df)
```

```python
#find average of the duration column in each run and record name of run
average_values = []
for df in data_frames:
     average_value = df['duration'].mean()
     average_values.append(average_value)
average_values
len(average_values)
    
```

```python
#average HFO rate
avg_hfo_rate_list = []
for df in data_frames:
    prev_time = 0
    num_hfo = 0
    avg_hfo_10s = 0
    for i in range(len(df.index)):   
        cur_time = math.trunc(df._get_value(i, 'onset')/10)
        if cur_time <= prev_time:           
            num_hfo += 1
        else:
            avg_hfo_10s += num_hfo
            #avg_hfo_10s
            prev_time = cur_time
            num_hfo = 0        
    avg_hfo_rate_list.append(math.ceil(avg_hfo_10s/30))
len(avg_hfo_rate_list)
```

```python
#brain areas variable
ripple_events = []
fast_ripple_events = []
fast_and_ripple_events = []
for df in data_frames:
    ripple = 0
    fast_ripple = 0
    fast_and_ripple = 0
    for i in range(len(df.index)):
        x = df.at[i, 'trial_type']
        if 'ripple' in x:
            ripple += 1
        elif 'fr' in x:
            fast_ripple += 1
        elif 'frandr' in x:
            fast_and_ripple += 1
            
    ripple_events.append(ripple)
    fast_ripple_events.append(fast_ripple)
    fast_and_ripple_events.append(fast_and_ripple)
    
    

            

len(ripple_events)
len(fast_ripple_events)
len(fast_and_ripple_events)





            
        
```
