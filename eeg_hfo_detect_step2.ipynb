```python

bids_path = BIDSPath(root = Path('../Untitled Folder/'), subject='01', session = 'interictalsleep', run = '01', datatype = 'ieeg', extension = '.vhdr')
bids_path
kwargs = {
    'filter_band': (80, 250),
    'threshold': 3,  # threshold for "significance"
    'win_size': 100,  # window size in samples
    'overlap': 0.25,
    'sfreq': 2000.0,
    # overlap in percentage relative to the window size
}

raw = read_raw_bids(bids_path)
```
