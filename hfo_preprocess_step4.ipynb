```python
final_data_frame = pd.read_csv('final_data_frame.csv')
final_data_frame_s = final_data_frame.iloc[:, 1:]
final_data_frame_s
```

```python
kms = KMeans(n_clusters=20, random_state=1).fit(final_data_frame_s[list(final_data_frame_s.iloc[:, :156])])
final_data_frame_s['Cluster'] = kms.labels_


```

```python
for i in range(156):
    final_data_frame_s[f'Cluster Centroid D{i}'] = final_data_frame_s['Cluster'].apply(lambda label: kms.cluster_centers_[label][i])
final_data_frame_s
```

```python
features = final_data_frame_s.columns.tolist()[0:156]
print(f"Features: \n{features}")


centroids = kms.cluster_centers_
print(f"Centroids \n{centroids}")
```

```python
sorted_centroid_features_idx = centroids.argsort(axis=1)[:,::-1]
print(f"Sorted Feature/Dimension Indexes for Each Centroid in Descending Order: \n{sorted_centroid_features_idx}")

print()

sorted_centroid_features_values = np.take_along_axis(centroids, sorted_centroid_features_idx, axis=1)
print(f"Sorted Feature/Dimension Values for Each Centroid in Descending Order: \n{sorted_centroid_features_values}")
```

```python
first_features_in_centroid_1 = centroids[0][sorted_centroid_features_idx[0]]
centroids
```

```python
final_data_frame_s.loc[[4,5], [""]] += final_data_frame_s.loc[[4,5], ["Feature1"]] + 30
```

```python
print(list(
        zip(
            [features[feature] for feature in sorted_centroid_features_idx[0]], 
            first_features_in_centroid_1
        )
    ))
```

```python
my_dict = {}
keys = []
values = []
for i in range(1, 20):
    keys.append(i)

for i in range(20):
    values.append(0)

for i in keys:
    my_dict[0] = 1
    my_dict[i] = values[i]

my_dict

print(final_data_frame_s['Cluster'].value_counts())
final_data_frame_s['Binary Cluster 0'] = final_data_frame_s['Cluster'].map(my_dict)
print("\n", final_data_frame_s["Binary Cluster 0"].value_counts())


    
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=1)
clf.fit(final_data_frame_s[(list(final_data_frame_s.iloc[:,:156]))].values, final_data_frame_s["Binary Cluster 0"].values)

# Index sort the most important features
sorted_feature_weight_idxes = np.argsort(clf.feature_importances_)[::-1] # Reverse sort

# Get the most important features names and weights
most_important_features = np.take_along_axis(
    np.array(final_data_frame_s.iloc[:, :].columns.tolist()), 
    sorted_feature_weight_idxes, axis=0)
most_important_weights = np.take_along_axis(
    np.array(clf.feature_importances_), 
    sorted_feature_weight_idxes, axis=0)

# Show
list(zip(most_important_features, most_important_weights))
```
